# Subplots

Exploring the matplotlib subplots OOP method of generating plots 

http://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
import matplotlib.pyplot as plt
import numpy as np

"""You can use subplots to create many plots. It takes a nrows and ncols argument to determine the subplots across each row and column
respectively. It returns a figure and an array of axes objects."""

x = plt.subplots(1,1)
```

Check the type of x 

```python
# x is a tuple containing a figure object and the axes subplot object
x
```

```python
type(x)
```

As we can see x is a tuple. The first element in the tuple contains the figure object, and the second element the axes subplot object. We can see the further below, by assigning the fig, ax elements like so.....

```python
# Tuple unpacking to assign the figure and subplots to seperate variables
fig, ax = plt.subplots(1,1)
```

```python
type(fig)
```

```python
type(ax)
```

```python
x = plt.subplots(1,2)
```

```python
#ax will be an array storing two axes objects elements
fig, ax = plt.subplots(1,2)
```

```python
# indexing to return the first axes object
ax[0]
```

```python
fig, ax = plt.subplots(1,2)
ax[0].set_title('First plot')
```

```python
#unpacking to assign the subplots to two seperate variables
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.set_title('First plot')
```

```python
#unpackaging to assign the subplots to four separate variables

fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2,2)

ax1.set_title('Ax1')
ax2.set_title('Ax2')
```
